# Types of arguments:-

```python
#1)Positional arguments

def display(a,b):
    print("a = ",a)
    print("b = ",b)
display(2,3)
```

```python
# 2)Default arguments
def display(a,b=10): #b is default value
    print("a = ",a)
    print("b = ",b)
display(2)
#############################
#o/p:
#a =  2
#b =  10
##############################
def display(a,b=10): #b is default value
    print("a = ",a)
    print("b = ",b)
display(2,3)
###########################
#o/p:
#a =  2
#b =  3
##### Positional pararmeter comes before default parameter
###################################
def display(a,b=10,c=11): #b is default value
    print("a = ",a)
    print("b = ",b)
display(2)

"""o/p:
a =  2
b =  10 """
```

```python
# 3)Keyword arguments
def display(a,b): #b is default value
    print("a = ",a)
    print("b = ",b)
display(b=2,a=3)

#positionwise value is not assign it is based on keywords
```

# *args

```python
# in c++ java MethodOverLoading :- same name of function but different number of arguments
# in python this type of methodoverloading is not support
# in python it is achive by *agrs(this is another option in python for method overloading)
```

```python
# *args it is use to accept positional arguments in the form of tuple
# * = multiple , args = arguments
# args can be list, tuple, dict, file
```

```python
def add(*args): #take multiple parameters
    sum=0
    for i in args:
        sum = sum+i
    print(sum)
add(2,3)
add(2,3,4)
add(2,3,4,5)
    
```

# *kwargs

```python
##  **kwargs in this kwargs it is not compalsoury we can take any variable name instead of kwargs
# in kwargs o/p values comes in form dictionary form
# **kwargs it accepts multiple keyword arguments in the form of dictionary
```

```python
def display(**kwargs):
    print(kwargs)
display(a=2,b=3)
```

```python
def display(**args):
    print(args)
display(a=2,b=3)
```
