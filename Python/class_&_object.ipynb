# Inheritance

```python
# 1)inherit the properties of one class ot another
# 2)reuseability of code
```

```python
# 1)Single Inheritance:-
class base:
    pass
class derived(base):
    pass
```

```python
class Person:
    def __init__(self,name,gender,age):
        self.name = name
        self.gender = gender
        self.age = age
    def display(self):
        print(self.name, self.gender, self.age)
```

```python
class Student(Person):
    def __init__(self, name, gender, age):
        Person.__init__(self, name, gender, age)
s1 = Student('Steve','M', 24)
s1.display()
```

```python
# what is diffefence between method overloading and method overriding ?
```

```python
class Student(Person):
    def __init__(self, name, gender, age,branch):
        Person.__init__(self, name, gender, age)
        self.branch = branch
s1 = Student('Steve','M', 24, 'computer')
s1.display()
```
