# Exception Handling

```python
#Types of error :-
# Syntax error
# semantic error (logical)
# Runtime (exception) - when code is executing then at that time this error is come
```

# Value Error :-

```python
# Value Error :-
# occurs when the type id correct but value is invlaid
# type is correct but invalid value
# eg:-
try:
    a = int("two")
    print(a)
except ValueError:
    print("invalid value for integer conversation")
```

```python
try:
    a = int("7")
    print(a)
except ValueError:
    print("invalid value for integer conversation")
```

# Type Error :-

```python
#occurs when operation is performed on incomplitable types
# when datatype is wrong
#eg:
try:
    result = "10"+5
    print(result)
except TypeError:
    print("unsupported operand types")
```

```python
try:
    result = "10"+"5"
    print(result)
except TypeError:
    print("unsupported operand types")
```

```python
try:
    result = 10+5
    print(result)
except TypeError:
    print("unsupported operand types")
```

# KeyError :-

```python
# KeyError :-
# occurs when accessing a non-existent key in a dictionary
```

```python
try :
    data = {'name':'rutuja'}
    print(data["age"])
except KeyError:
    print("key not found")
```

```python
try :
    data = {'name':'rutuja'}
    print(data["name"])
except KeyError:
    print("key not found")
```

# Generic Exception Handling

```python
try :
    x = int("abc")
except Exception as e:
    print("Error occured : ",e)
```

# else block:-

```python
# Execute only if no exception occur
# if all code run successfully then else part run
try:
    num = int("10")
except ValueError:
    print("error")
else:
    print("conversion successful")
```

# Finally :-

```python
#this block always execute
```

```python
try:
    num = int("10")
except ValueError:
    print("error")
else:
    print("conversion successful")
finally:
    print("end of all code")
```
