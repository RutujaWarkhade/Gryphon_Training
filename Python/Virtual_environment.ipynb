# Creating VE (using venv)

```python
# Step 1: create python version  
###  python --

# Step 2: create virtual environment
###  python -m venv myenv
#This create a folder
#  myenv/
#   |- Scripts/(for windows OS)
#   |- bin (for linux/mac OS)
#   |-Lib
#   |-Include
#   |-pyvenv.cfg

# Step 3:- Activate the VE
#### myenv\Scripts\activate
# if want to deactivate VE
### myenv\Scripts\deactivate
```
