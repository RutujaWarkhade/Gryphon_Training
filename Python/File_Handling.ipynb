# File Handlling 

#Operation:-

1)open

2)close

3)read

4)write

5)Append

```python
# 1) open :-
fd = open('sample.txt','r') 
print(fd)
```

```python
# fd points to the 1st byte of the file
```

```python
# 2) close :-
fd = open('sample.txt','r') 
fd.close()
```

```python
# fd.close() by this file discreptor get deleted
```

```python
# 3)read :-
fd = open('sample.txt','r') 
print(fd.read())
fd.close()

#################

fd = open('sample.txt','r') 
print(fd.read(4)) #no. of bytes (chars) to read
fd.close()
```

```python
# file pointer moves foeward by the no. of bytes read
fd = open('sample.txt','r') 
print(fd.read(4)) 

print(fd.read(4)) 
```

```python
# 4) readline
# it read 1 line at a time
fd = open('sample.txt','r') 
line = fd.readline()
print(line) 
fd.close()
```

```python
# 5) readlines
# it read 1 line at a time
fd = open('sample.txt','r') 
line = fd.readlines()
print(line) 
fd.close()
```

```python
# 6)write
# if file exist first and in that file if we write again the that file should overright (means clean 1st data then write new)
# in write mode writing mode starts from startinh=g
fd = open('rutuja.txt','w')
fd.write('its a new line')
fd.close()
```

```python
fd = open('rutuja.txt','r') 
line = fd.readlines()
print(line) 
fd.close()
```

```python
fd = open('rutuja.txt','w')
fd.writelines(['this is line1\n','this is line 2'])
fd.close()
```

```python
# 7)append
# in this mode writing of file starts from ending
fd = open('rutuja.txt','a')
fd.writelines(['\nthis is line1\n','this is line 2'])
fd.close()
```

```python
fd = open('rutuja.txt','r') 
line = fd.readlines()
print(line) 
fd.close()
```
